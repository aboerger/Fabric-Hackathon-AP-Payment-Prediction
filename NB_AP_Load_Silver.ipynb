```python
import re
from notebookutils import mssparkutils

def load_to_delta(table_name):
    spark.conf.set("sprk.sql.parquet.vorder.enabled", "true")
    spark.conf.set("spark.microsoft.delta.optimizeWrite.enabled", "true")
    path = "abfss://6c54ba89-e065-4ece-84e8-ee42ca7bed9e@onelake.dfs.fabric.microsoft.com/629b64b0-d670-4c1f-b46d-23d68d35d841/Tables/" #LH_AP_Silver
    sdf = spark.read.parquet("Files/ADI/" + table_name)
    pattern = r'[ ,;{}\(\)\n\t=]'  # Regex pattern for prohibited characters in Delta column names
    sdf = sdf.toDF(*[re.sub(pattern, '', x) for x in sdf.columns])
    sdf.write.mode("overwrite").option("mergeSchema","true").format("delta").save(path + table_name)
```

```python
folders = mssparkutils.fs.ls('Files/ADI')
for folder in folders:
    load_to_delta(folder.name)
    print(folder.name)
```
